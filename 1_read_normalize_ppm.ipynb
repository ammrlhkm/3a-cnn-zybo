```python
import numpy as np
import cv2
from matplotlib import pylab as plt
```

```python
filename="image1.ppm"

img = cv2.imread(filename)
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```
